# Q6
How similar are two artists? 

```python
user_artists_with_name
```

```python
def artist_sim(aid1, aid2):
    # first check aid1 in the list
    if aid1 in user_artists_with_name['artistID']:
        # second check aid2 in the list
        if aid2 in user_artists_with_name['artistID']:
            # if both aid1 and aid2 in the list, add userID in the set
            aid1_set = set(user_artists_with_name[user_artists_with_name['artistID'] == aid1]['userID'])
            aid2_set = set(user_artists_with_name[user_artists_with_name['artistID'] == aid2]['userID'])
            # compute intersection and union of aid1 and aid2
            intersection = len(list(set(aid1_set).intersection(aid2_set)))
            union = (len(aid1_set) + len(aid2_set)) - intersection
            # I use iloc[1] to output only one name of aid1 and aid2
            print(user_artists_with_name[user_artists_with_name['artistID'] == aid1]['name'].iloc[1],
            "/",user_artists_with_name[user_artists_with_name['artistID'] == aid2]['name'].iloc[1])
            # print similarity
            print(float(intersection) / union)
        # if aid2 does not in the list, print 
        else:
            print(aid2, ' does not in the list')
    # if aid1 does not in the list, print
    else:
        print(aid1, ' does not in the list')
```

```python
artist_sim(735, 562)
artist_sim(735,89)
artist_sim(735,289)
artist_sim(89,289)
artist_sim(89,67)
artist_sim(67,735)
```

# Q7


 Recommend artists based on friends

```python
def artist_rec(uid):
    # Reset the index for user_friends
    uf = user_friends.reset_index()
    # Reset the index for user_artists_df
    ua = user_artists.reset_index()
    # Find user's friendID
    a1 = uf[uf['userID'] == uid]['friendID']
    #Create an empty list
    a2 =[]
    # append artistID listened by users' friends
    for friend in a1:
        a2.append(ua[ua['userID'] == friend][['artistID','weight']])
    a3 = pd.concat(a2)
    # count number of friend who listedned the same artist
    counts = a3.artistID.value_counts()
    # find sum of weight for each artist
    a4 = a3[a3['artistID'].isin(counts.index[counts > 1])]
    # find sum of weight for each artist
    a4_sum = a4.groupby('artistID').sum()
    # find count of friends listeners for each artist
    a4_count = a4.groupby('artistID').count()
    # Merge previous sum and count
    p7 = pd.merge(a4_sum, a4_count, left_index = True, right_index = True)
    # compute the average number of plays of each artist by all friends
    p7['average'] = round(p7['weight_x'] / p7['weight_y'])
    # Merge the artists_df to find corresponding name 
    p7 = pd.merge(p7,artists['name'],left_index = True, right_index = True).reset_index()
    # Change the columns name to easily understand
    p7 = p7.rename(columns = {'index': 'artistID', 'weight_x':'tital_plays',
                             'weight_y':'num_listeners'})
    # Subset the dataframe not include the artist which user have already listened it 
    # and selct the top 10 by average
    p7 = p7[~p7['artistID'].isin(ua[ua['userID'] == 2]['artistID'])] \
        .sort_values('average',ascending= False).head(10).reset_index(drop=True)
    return p7[['artistID', 'name']]
```
